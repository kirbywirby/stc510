Adding columns year and month to dataframe. Using DatetimeIndex to get the month and year from column 'OCCURRED ON'. 

I found this on Stack Overflow and was worried that it wouldn't work. Really excited that it did! 

```python
n['OCCURRED ON'].describe
```

```python
n['year'] = pd.DatetimeIndex(n['OCCURRED ON']).year
n['month'] = pd.DatetimeIndex(n['OCCURRED ON']).month
```

Adding column 'CRIME TYPE' for non-violent and violent crimes

```python
n['CRIME TYPE'] = ''
```

```python
n.loc[n['UCR CRIME CATEGORY'] == 'LARCENY-THEFT', 'CRIME TYPE'] = 'non-violent'
n.loc[n['UCR CRIME CATEGORY'] == 'BURGLARY', 'CRIME TYPE'] = 'non-violent'
n.loc[n['UCR CRIME CATEGORY'] == 'MOTOR VEHICLE THEFT',  'CRIME TYPE'] = 'non-violent'
n.loc[n['UCR CRIME CATEGORY'] == 'DRUG OFFENSE',  'CRIME TYPE'] = 'non-violent'
n.loc[n['UCR CRIME CATEGORY'] == 'ARSON', 'CRIME TYPE'] = 'non-violent'

n.loc[n['UCR CRIME CATEGORY'] == 'AGGRAVATED ASSAULT', 'CRIME TYPE']='violent'
n.loc[n['UCR CRIME CATEGORY'] == 'ROBBERY', 'CRIME TYPE']='violent'
n.loc[n['UCR CRIME CATEGORY'] == 'RAPE', 'CRIME TYPE']='violent'
n.loc[n['UCR CRIME CATEGORY'] ==  'MURDER AND NON-NEGLIGENT MANSLAUGHTER', 'CRIME TYPE']='violent'
```

Newly joined data set

```python
nmeans1.loc[85053]
```

This is a bit messy but I am creating 2 new columns- Non-violent count and violent count. I wanted these counts to be separate columns instead of mashed together under the "count" column.

```python
pd.options.mode.copy_on_write = True
nmeans1['non-violent count'] = ''
```

```python
pd.options.mode.copy_on_write = True
nmeans1['non-violent count']=nmeans1['count'][nmeans1['CRIME TYPE']=='non-violent']
```

```python
nmeans1['violent count']=nmeans1['count'][nmeans1['CRIME TYPE']=='violent']
```

Resetting index so I can remove duplicate zip codes

```python
nmeans1.reset_index(inplace = True)
```

Creating nmeans2 - de-duped dataset

```python
nmeans2= nmeans1.drop_duplicates(subset = 'ZIP')
```

Removing columns I don't need anymore

```python
nmeans2.drop('count', axis = 1, inplace = True)
```

```python
nmeans2.drop('CRIME TYPE', axis = 1, inplace = True)
```

```python
nmeans2
```

```python
nmeans2.set_index('ZIP', inplace = True)
```

New, cleaned data set showing mean income, violent crimes, and non-violent crimes by zip code!

```python
nmeans2
```

Creating scatter plot showing both non-violent and violent crimes by mean income for the top 20 zip codes.

```python
fig = plt.figure()
ax1 = fig.add_subplot()

ax1.scatter(x = nmeans2['Mean'] , y = nmeans2['violent count'], s=10, c='b', marker="s", label='violent crimes')
ax1.scatter(x = nmeans2['Mean'],y = nmeans2['non-violent count'], s=10, c='r', marker="o", label='non-violent crimes')
plt.legend(loc='upper right')
ax1.set_title('Violent and Non-Violent Crime by Mean Income in Top 20 Phoenix Zip Codes')
ax1.set_xlabel('Mean Income')
ax1.set_ylabel('Crime #s')
plt.show()
```

Scatter plot showing only violent crimes by mean income
